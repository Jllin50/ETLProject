```python
import pandas as pd
import csv
from sqlalchemy import create_engine
```

```python
countries = "countries.csv"
countries_db = pd.read_csv(countries)
```

```python
countries_db = countries_db.dropna()
```

```python
economy = "efw_cc.csv"
economy_db = pd.read_csv(economy)
```

```python
economy_db = economy_db.dropna()
```

```python
economy_db.dtypes
```

```python
countries_db = countries_db.rename(columns={"Country": "countries"})
```

```python
countries_db.dtypes
```

```python
countries_db = countries_db.set_index('countries')
```

```python
economy_db = economy_db.set_index('countries')
```

```python
merged_db = countries_db.join(economy_db, on='countries', how='outer')
```

```python
merged_db = merged_db.rename(columns={"Area (sq. mi.)": "Area", "Pop. Density (per sq. mi.)": "Pop_Density", "Coastline (coast/area ratio)": "Coast_to_area_ratio", "Infant mortality (per 1000 births)": "Infant_Mortality", "GDP ($ per capita)":"GDP", "Literacy (%)": "Literacy_Percentage", "Phones (per 1000)": "Phones_per_1000", "Arable (%)":"Arable_Percent", "Crops (%)": "Crops", "Other (%)": "Other" })
```

```python

rds_connection_string = "postgres:237096@localhost:5432/postgres"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
merged_db.to_sql('EconomyandCountry', engine)
```
