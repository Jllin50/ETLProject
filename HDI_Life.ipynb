```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
human_dev_file = "C:\\Users\\Spring\\Desktop\\world project\\Resources\\Human_Development_Index.csv"
human_dev_df = pd.read_csv(human_dev_file)
human_dev_df.head()
```

```python
human_dev_cols = ["HDI Rank", "Country", "2016"]
human_dev_transformed= human_dev_df[human_dev_cols].copy()
human_dev_transformed.head()
```

```python
human_dev_transformed = human_dev_transformed.rename(columns={'HumanDevelopmentIndex':'HDI Rank','2016':'HumanDevelopmentIndex'})
human_dev_transformed.head()
```

```python
life_exp_file = "C:\\Users\\Spring\\Desktop\\world project\\Resources\\indicator-life_expectancy_at_birth.csv"
life_exp_df = pd.read_csv(life_exp_file)
life_exp_df.head()
```

```python
life_exp_cols = ["country", "2016"]
life_exp_transformed= life_exp_df[life_exp_cols].copy()
life_exp_transformed = life_exp_transformed.rename(columns={'country':'Country','2016':'Life Expectancy'})
```

```python
life_exp_transformed.head()
```
